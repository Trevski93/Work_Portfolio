## NBA Page Rank Algorithm

```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plot
import csv
plot.rc('font',family='serif')
#plot.rc('text', usetex=True)
plot.rc('xtick',labelsize=14)
#matplotlib.rc('text.latex')
```

```python
# There are 30 NBA Teams
nbTeams=30
# winLoss[i,j] = nb of wins of team i against team j
winLoss = np.zeros((nbTeams,nbTeams))
teams={}
print(teams)
```

```python
# Reading data: constructing Win/Loss matrix
with open('nba17-18.csv', 'r') as nbaData:
    reader = csv.reader(nbaData, delimiter=',')
    nbTeam=0
    for row in reader:
        team1=row[0]
        if team1 in teams:
            t1 = teams[team1]
        else:
            t1 = nbTeam
            teams[team1]=t1
            nbTeam+=1      
        score1=int(row[1]) # Score of the team1
        team2=row[2]
        if team2 in teams:
            t2 = teams[team2]
        else:
            t2 = nbTeam
            teams[team2]=t2
            nbTeam+=1
        score2=int(row[3]) # Score of the team2

        if score1 > score2:
            winLoss[t1,t2]+=1
        else:
            winLoss[t2,t1]+=1

# Constructing team list
team=[0]*nbTeams
for p,i in teams.items():
    team[i]=p
print(len(team))
```

```python
# Wins and Loss of each team
win = np.sum(winLoss,axis = 1)
loss = np.sum(winLoss,axis = 0)
```

```python
# Ranking teams according to their percentage of wins
ratio=win/(win+loss)
sort = np.flip(np.argsort(ratio),0)
pr = np.around(100*ratio,decimals=1)
for i in range(nbTeams):
    print(team[sort[i]] + ' (' + str(pr[sort[i]]) + '%)')
```

```python
team= [team[i] for i in sort]
ratio = np.array([ratio[i] for i in sort])
```

## Ranking of NBA Teams by Win Percentage

```python
# Plot ranking

y=-np.array(range(nbTeams))
plot.figure(figsize=(12,25),frameon=False)
plot.barh(y,100*ratio,color='aquamarine', height=0.9, edgecolor = 'black',linewidth =0.5)
for i in range(nbTeams):
    lb = r'\Huge '
    lb = lb + team[i]
    plot.text(0.0922,y[i]-0.35,lb)
t=plot.yticks([],[])
l=plot.ylim(-nbTeams+ 0.4,0.6)
ax = plot.gca()
ax.xaxis.tick_top()
plot.savefig("ratio_nba.pdf",bbox_inches='tight',transparent=True)
```

```python
#Create Transition Matrix 

p = np.zeros((winLoss.shape[0],winLoss.shape[1]))

for i in range(0,p.shape[0]):
    for j in range(0,p.shape[1]):
        
        if i == j: 
            p[i][j] = sum(winLoss[i,:])/ (sum(winLoss[i,:]) + sum(winLoss[:,i]))
            
        else:
            p[i][j] = winLoss[i][j] / (sum(winLoss[j,:]) + sum(winLoss[:,j]))
            
```

```python
#Transition Matrix verification
for i in range(0,p.shape[0]):
    print(sum(p[:,i]))
```

```python
eigenvalues, eigenvectors = np.linalg.eig(p)
print(list(eigenvalues)[0])
mu_complex = eigenvectors[:,0]*(-1)
mu = [i.real for i in mu_complex]
print(mu)
```

```python
#Returns the indices that would sort an array and then reverses the order
sort = np.flip(np.argsort(mu),0)
#round the sorted values
pr = np.around(mu*100,decimals=50)
for i in range(nbTeams):
    print(team[sort[i]] + ' (' + str(pr[sort[i]]) + '%)')
```

## NBA Ranking by Google Page Rank Algorithm

```python
team = [team[i] for i in sort]
mu = np.array([mu[i] for i in sort])

# Plot ranking

y=-np.array(range(nbTeams))
plot.figure(figsize=(12,25),frameon=False)
plot.barh(y,100*mu,color='aquamarine', height=0.9, edgecolor = 'black',linewidth =0.5)
for i in range(nbTeams):
    lb = r'\Huge '
    lb = lb + team[i]
    plot.text(0.0922,y[i]-0.35,lb)
t=plot.yticks([],[])
l=plot.ylim(-nbTeams+ 0.4,0.6)
ax = plot.gca()
ax.xaxis.tick_top()
plot.savefig("ratio_nba.pdf",bbox_inches='tight',transparent=True)
```
